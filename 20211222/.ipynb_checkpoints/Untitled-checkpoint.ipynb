```python
%matplotlib inline



import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '../data/crime.csv' 
crimes = pd.read_csv(file_path, encoding='cp949')
crimes = crimes.fillna(0)
```

```python
print(crimes.info())
```

```python
crimes
```

```python
# 수치형 피처의 각 컬럼별 요약 통계(갯수, 평균, 표준편차, 최솟값, 4분위수, 최댓값) 지표 출력
crimes.describe()
```

```python
result = crimes.groupby('구분').sum()
result
```

```python
x_name=result.index.tolist()
x_name
```

```python
plt.rc('font', family='NanumGothic')

n_groups = len(result.index)
a = result.loc['마약류관리법'].tolist()
b = result.loc['성폭력처벌법'].tolist()
c = result.loc['아동청소년보호법'].tolist()
d = result.loc['여객자동차운수사업법'].tolist()
e = result.loc['특정강력범죄처벌법'].tolist()
f = result.loc['특정범죄가중처벌법'].tolist()
g = result.loc['형법상습'].tolist()



index = np.arange(7)
bar_width = 0.1

rects1 = plt.bar(index, a, bar_width, color='r',label='마약류관리법')
rects2 = plt.bar(index + bar_width, b, bar_width, color='g',label='성폭력처벌법')
rects3 = plt.bar(index + bar_width * 2, c, bar_width, color='b',label='아동청소년보호법')
rects4 = plt.bar(index + bar_width * 3, d, bar_width, color='black',label='여객자동차운수사업법')
rects5 = plt.bar(index + bar_width * 4, e, bar_width, color='skyblue',label='특정강력범죄처벌법')
rects6 = plt.bar(index + bar_width * 5, f, bar_width, color='pink',label='특정범죄가중처벌법')
rects7 = plt.bar(index + bar_width * 6, g, bar_width, color='orange',label='형법상습')

plt.title('년도별 범죄')
plt.xticks(index, result.columns)
plt.legend() 
plt.show()
```
